```python
df = pd.read_csv('/Users/brianoktavec/MSC550/Assignment_2/election2000.csv')
```

```python
# 1.1
```

```python
import statsmodels.api as sm
ols = sm.OLS(df.iloc[:, 11], sm.add_constant(df.iloc[:, 1:11]))
lm = ols.fit()
print(lm.summary())
```

```python
# 1.2
```

```python
import scipy.stats as st
from scipy.stats import ttest_ind
```

```python
for x in df:
    print(x + " Test:")
    y = ttest_ind(df['Bush'], df[x])
    print(y)
    print("")
```

```python
# 1.3
```

```python
fitted_y = lm.fittedvalues
residual = lm.resid
plt.figure(figsize=(12,6))
plt.scatter(fitted_y,residual)
plt.title('Residuals v.s. Fitted')
plt.xlabel('Fitted values')
plt.ylabel('Residuals')
```

```python
from sklearn.preprocessing import scale
residual_norm = scale(residual)
sm.qqplot(residual_norm, line='45')
plt.figure(figsize=(12,6))
plt.hist(x=residual_norm, bins='auto', rwidth=0.85)
plt.title('Histogram for standardized residuals')
plt.xlabel('Standardized residuals')
plt.ylabel('Frequency')
```

```python
# 1.4
```

```python
from scipy import stats
from scipy.stats import kstest
print(stats.shapiro(residual))
print(kstest(residual, 'norm'))
```
